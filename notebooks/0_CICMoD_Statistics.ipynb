### Statistics for indices included in CICMoD data set

In this project we work with **C**limate **I**ndex **C**ollection based on **Mo**del **D**ata (CICMoD) data set (https://github.com/MarcoLandtHayen/climate_index_collection). Here, we show some statistics:

- Pairwise correlation coefficients
- Mean and standard deviation

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

### Load data

Load collection of climate indices directly from GitHub release.

```python
data_url = (
    "https://github.com/MarcoLandtHayen/climate_index_collection/"
    "releases/download/v2022.09.09.1/climate_indices.csv"
)
```

```python
climind = pd.read_csv(data_url)
climind = climind.set_index(["model", "year", "month", "index"]).unstack(level=-1)["value"]
climind
```

```python
# Plot pairwise correlation coefficients (FOCI):
fig, ax = plt.subplots(figsize=(9,7))
plt.title('Pairwise correlation of climate indices (FOCI)', fontsize=18)
sns.heatmap(climind.loc["FOCI"].corr(), cmap="RdBu", xticklabels=True, yticklabels=True)
```

```python
# Plot pairwise correlation coefficients (CESM):
fig, ax = plt.subplots(figsize=(9,7))
plt.title('Pairwise correlation of climate indices (CESM)', fontsize=18)
sns.heatmap(climind.loc["CESM"].corr(), cmap="RdBu", xticklabels=True, yticklabels=True)
```

```python
# Compute mean and standard deviation (FOCI):
print("FOCI mean:\n==========\n",climind.loc["FOCI"].mean())
print("\n\nFOCI std dev:\n=============\n",climind.loc["FOCI"].std())
```

```python
# Compute mean and standard deviation (CESM):
print("CESM mean:\n==========\n",climind.loc["CESM"].mean())
print("\n\CESM std dev:\n=============\n",climind.loc["CESM"].std())
```
